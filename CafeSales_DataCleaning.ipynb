```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# load the dataset
df=pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/dirty_cafe_sales.csv")
df
```

```python
#  Rename columns
d = {
    "Transaction ID": "transaction_id",
    "Price Per Unit": "price (per_unit)",
    "Transaction Date": "transaction_date"
}

df.rename(columns=d, inplace=True)
df.head()
```

```python
#Drop Duplicate Rows
df.drop_duplicates(inplace=True)
df
```

```python
#Handling Missing Values
df.isnull().sum()
```

```python
print("Unique values in 'Item' column:", df['Item'].unique())
print("Unique values in 'Quantity' column:", df['Quantity'].unique())
print("Unique values in 'Total Spent' column:", df['Total Spent'].unique())
print("Unique values in 'Payment Method' column:", df['Payment Method'].unique())
print("Unique values in 'Location' column:", df['Location'].unique())
```

```python
df.replace(['UNKNOWN', 'ERROR'], np.nan, inplace=True)

print("Updated unique values in 'Item' column:", df['Item'].unique())
print("Updated unique values in 'Quantity' column:", df['Quantity'].unique())
print("Updated unique values in 'Total Spent' column:", df['Total Spent'].unique())
print("Updated unique values in 'Payment Method' column:", df['Payment Method'].unique())
print("Updated unique values in 'Location' column:", df['Location'].unique())

df.isnull().sum()
```

```python
# Drop Missing Rows
df.dropna(inplace=True)
df
```

```python
# Dropping Unwanted Columns
df = df.drop(["transaction_id"], axis=1)

df.head()
```

```python
#Dropping Unwanted Rows (By Index)
df = df.drop([0, 1])   # dropping first two rows
df.head()
```

```python
#Resetting Index
df.reset_index(drop=True, inplace=True)
df.head()
```
